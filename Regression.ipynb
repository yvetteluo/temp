```python
# Loading modules
import pandas as pd
import statsmodels.formula.api as smf
# Import plotting module
import seaborn as sns
import numpy as np
# import necessary library
from stargazer.stargazer import Stargazer
```

# Weekly data

```python
weekly = pd.read_csv('weekly_final_data.csv')
```

```python
weekly.head(2)
```

```python
weekly.shape
```

```python
for i in weekly.columns:
    b = i.lower()
    b = b.replace('*','')
    b = b.replace('/','')
    weekly = weekly.rename(columns={i:b})
```

```python
weekly.columns
```

```python
#Creating a dummy for treatment
weekly['isinvc'] = [1 if i == 'Yes' else 0 for i in weekly.isinvc]
```

```python
#Defining a list of dependent variables
DV_1w = ['marketcap_avg1w','price_avg1w','pricevolume_avg1w',
         'percent_change_7d_avg1w','volume_24h_avg1w']

# 'pricevolumemarketcap_avg1w'
```

```python
mod1 = smf.ols(formula= 'volume_24h_avg1w ~ isinvc', data = weekly).fit(cov_type='HC0') 
```

```python
mod1.summary()
```

```python
models = []
for var in DV_1w:
    
    # Define the model
    mod = smf.ols(formula=f'{var} ~ isinvc', data = weekly).fit(cov_type='HC0') 

    models.append(mod)
    
#Looking at the results
table_result = Stargazer(models)

#Adding the variable names to each model
table_result.custom_columns(DV_1w, [1 for i in range(len(DV_1w))])
table_result
```

```python
weekly.ico_price
```

```python
models = []
for var in DV_1w:
    
    # Define the model
    mod = smf.ols(formula=f'{var} ~ isinvc+category+country+funding_date+platform+type+amount_raised+hardcap+softcap+kyc+token_supply+whitelist+percent_of_hardcap+whitepaper+pre_start_date+pre_end_date+token_for_sale+ico_price', data = weekly).fit(cov_type='HC0') 

    models.append(mod)
    
#Looking at the results
table_result = Stargazer(models)

#Adding the variable names to each model
table_result.custom_columns(DV_1w, [1 for i in range(len(DV_1w))])
table_result
```
